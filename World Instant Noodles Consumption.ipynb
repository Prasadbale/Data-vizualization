```python
#Setting country as index

data.head()
```

```python
data_sorted = data.sort_values(['2022','2021','2020','2019','2018'],ascending=False)
```

```python
plt.figure(figsize=(14,8))
sns.lineplot(data=data_sorted,y='2018',x='Country/Region',marker='o',label='2018')
sns.lineplot(data=data_sorted,y='2019',x='Country/Region',marker='o',label='2019')
sns.lineplot(data=data_sorted,y='2020',x='Country/Region',marker='o',label='2020')
sns.lineplot(data=data_sorted,y='2021',x='Country/Region',marker='o',label='2021')
sns.lineplot(data=data_sorted,y='2022',x='Country/Region',marker='o',label='2022')
plt.legend()
plt.ylabel('Consumption(in $M)')
plt.xticks(rotation=90)
plt.show()
```

We can see that the cosumption is following same pattern every year across the countries with China, Indonesian and India as top 3 Consumers of Instant Noodles

```python
#Consumption of top 5 countries over the years

df = data_sorted[['Country/Region','2018','2019','2020','2021','2022']][:5]
plt.figure(figsize=(14,12))
df.plot(kind='bar',stacked=True)

ticks = np.arange(0,5)
labels = data['Country/Region'][:5].values

plt.xticks(ticks=ticks,labels=labels)
plt.ylabel('Consumption',fontdict={'fontsize':9})
plt.title('Consumption of top 5 Countries over the years',fontdict={'color':'Brown','fontsize':10})
plt.show()
```

We can see that the YOY change remains almost the same across the top 5 countries

```python
data_sorted.head()
```

```python
# YOY changes

data['YoY_2019'] = round(((data['2019']-data['2018'])/(data['2018']))*100,2)
data['YoY_2020'] = round(((data['2020']-data['2019'])/(data['2019']))*100,2)
data['YoY_2021'] = round(((data['2021']-data['2020'])/(data['2020']))*100,2)
data['YoY_2022'] = round(((data['2022']-data['2021'])/(data['2021']))*100,2)
```

```python
data.head()
```

```python
data_sorted = data.sort_values(['2022','2021','2020','2019','2018'],ascending=False)
plt.figure(figsize=(12,6))
sns.lineplot(data=data_sorted,y='YoY_2019',x='Country/Region',marker='o',label='YoY_2019')
sns.lineplot(data=data_sorted,y='YoY_2020',x='Country/Region',marker='o',label='YoY_2020')
sns.lineplot(data=data_sorted,y='YoY_2021',x='Country/Region',marker='o',label='YoY_2021')
sns.lineplot(data=data_sorted,y='YoY_2022',x='Country/Region',marker='o',label='YoY_2022')
plt.legend()
plt.ylabel('Consumption(in $M)')
plt.xticks(rotation=90)
plt.show()
```
